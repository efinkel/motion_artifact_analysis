```python

task_data_files = np.array(glob.glob("Log_EF*"))
column_names1 =['mouse_name', 'date','block_type', 'trial_type', 'touch_stimulus',
                'vis_stimulus', 'response','trial_num', 'trial_onset', 'stim_onset', 'stim_offset', 
                'licks_right', 'licks_left', 'spike_times', 'cluster_name']

log_df = pd.DataFrame([], columns = column_names1)

for file_num in tnrange(len(task_data_files)):
    mat = sp.io.loadmat(task_data_files[file_num])
    log = mat['log']
    indv_log_df = pd.DataFrame(log, columns = column_names1)
    log_df = pd.concat([log_df,indv_log_df])
    
for col in [0,1,2,3,4,5,6,6,7,7,7,7,8,8,8,8,9,9,9,9,10,10,10,10,11,11,12,12,14]:
    log_df.iloc[:,col] = log_df.iloc[:,col].str[0]
    
log_df['stim_onset'] = log_df['stim_onset'].fillna(0)
log_df['spike_times(stim_aligned)'] = log_df['spike_times'] - log_df['stim_onset']
log_df['spike_times(stim_aligned)'] = log_df['spike_times(stim_aligned)'].apply(lambda x: np.concatenate(x) if len(list(x)) > 0 else x)
```

```python
log_df['uni_id'] = (log_df['mouse_name'].apply(lambda x: x[-3:]) + log_df['date']+log_df['cluster_name'].apply(lambda x: x[2] + x[-2:]))
unit_key_df = log_df[['uni_id', 'mouse_name', 'date', 'cluster_name']].drop_duplicates().reset_index(drop = True)

log_df['licks_right'] = log_df['licks_right'].apply(lambda y: y.ravel()) - log_df['stim_onset']
log_df['licks_left'] = log_df['licks_left'].apply(lambda y: y.ravel()) - log_df['stim_onset']
licks = log_df.loc[:,['licks_right', 'licks_left']].apply(lambda x: [np.concatenate(x.as_matrix())], axis = 1)
licks = licks.apply(lambda y: y[0])

log_df['first_lick'] =  licks.apply(lambda y: np.nan if len(y[y>0.1])==0 else min(y[y>0.1]))
```

```python
def unit_gen(log_df, motion_df):
    for unit in log_df['uni_id'].unique():
        unit_df = log_df[log_df['uni_id'] == unit]
        unit_df = unit_df.sort_values(['mouse_name', 'date', 'trial_type', 'first_lick']).iloc[::-1]
        unit_df = unit_df.reset_index(drop = True)
        
        date = unit_df.loc[0, 'date']
        ex_perch = motion_df.loc[motion_df['date'] == date, ['trial_num', 'perch1(-1 - 2)', 'perch2(-1 - 2)']]
        ex_perch = motion_df.loc[motion_df['date'] == date, ['trial_num', 'perch1(-1 - 2)', 'perch2(-1 - 2)']]
        ex_perch = unit_df.reindex(columns = ['date','trial_type', 'trial_num']).merge(ex_perch, how = 'left', on = 'trial_num')
        
        yield unit_df, ex_perch
```

```python

def process_to_plot(unit_df, ex_perch, window):
    w_inds =  (window+1)*30000
    touch_rows = ex_perch['trial_type'].str.contains('Som')

    data = np.vstack(ex_perch.loc[touch_rows, 'perch1(-1 - 2)'].as_matrix())
    data1 = np.vstack(ex_perch.loc[~touch_rows, 'perch2(-1 - 2)'].as_matrix())

    touch_all_perch = data[:, int(w_inds[0]):int(w_inds[1])]
    visual_all_perch = data1[:, int(w_inds[0]):int(w_inds[1])]

    touch_lick_ind = (unit_df.loc[touch_rows,'response'] == 1)
    visual_lick_ind = (unit_df.loc[~touch_rows,'response'] == 2)

    return touch_all_perch, visual_all_perch, touch_lick_ind.as_matrix(), visual_lick_ind.as_matrix()
```

```python
def plot_raster(unit_df, window, trial_type = 'Stim_Som'):
    trial_total = 0
    bin_size = 0.025
    spike_counts = []
    mod_df = unit_df[unit_df['trial_type'] == trial_type][::-1]
    for trial, spike in enumerate(mod_df['spike_times(stim_aligned)']):
        spike = spike[(spike>window[0]) & (spike<=window[1])]
        plt.vlines(spike, trial + trial_total + .5, trial + trial_total + 1.5)
        plt.vlines(mod_df.iloc[trial]['first_lick'], trial + trial_total + .5, trial +
                   trial_total + 1.5, color = 'C2', linewidth = 5)

        spike = spike[(spike>window[0]) & (spike<=window[1])]
        edges = np.arange(window[0], window[1]+bin_size*2, bin_size)
        count, _ = np.histogram(spike,edges)
        spike_counts.append(count)

    return edges, np.array(spike_counts)
```

```python
import matplotlib.gridspec as gridspec

def plot_perch(unit_df, ex_perch, window):
    
    touch_all_perch, vis_all_perch, touch_lick_ind, visual_lick_ind = process_to_plot(unit_df, ex_perch, window)
    
    touch_lick_perch = touch_all_perch[touch_lick_ind]
    visual_lick_perch = vis_all_perch[visual_lick_ind]
    
    fig = plt.figure()
    gs = gridspec.GridSpec(4, 2,
                           width_ratios=[1,1],
                           height_ratios=[2,2,1,1]
                           )

    ax= plt.subplot(gs[0])
    ax1 = plt.subplot(gs[1])
    ax2 = plt.subplot(gs[2])
    ax3 = plt.subplot(gs[3])
    ax4 = plt.subplot(gs[4])
    ax5 = plt.subplot(gs[5], sharey = ax4)
    ax6 = plt.subplot(gs[6])
    ax7 = plt.subplot(gs[7], sharey = ax6)
    axes = [ax,ax1,ax2,ax3,ax4,ax5,ax6,ax7]

    size_mult = 1
    bin_size = 0.025
    fig.set_figheight(14)
    fig.set_figwidth(10)

    vmin = np.min(np.concatenate([touch_all_perch, vis_all_perch]))
    vmax = np.max(np.concatenate([touch_all_perch, vis_all_perch]))

    # ax.imshow(data, aspect = 800, cmap = cmocean.cm.balance, vmin = vmin, vmax = vmax)
    im1 = ax.imshow(touch_all_perch, aspect = 'auto', cmap = cmocean.cm.balance, vmin = vmin, vmax = vmax, extent=[window[0],window[1],1, touch_all_perch.shape[0]])
    im2 = ax1.imshow(vis_all_perch, aspect = 'auto', cmap = cmocean.cm.balance, vmin = vmin, vmax = vmax, extent=[window[0],window[1],1, vis_all_perch.shape[0]])

    # ax.set_position([0.05,0.6,0.35, 0.35])

    plt.sca(ax2)
    edges, spike_counts_touch = plot_raster(unit_df, window)
    plt.sca(ax3)
    edges, spike_counts_vis = plot_raster(unit_df, window,trial_type = 'Stim_Vis')
    ax2.set_xlim(window); ax2.set_ylim([0,touch_all_perch.shape[0]])
    ax3.set_xlim(window); ax3.set_ylim([0,vis_all_perch.shape[0]])

  
    ax4.plot(np.arange(window[0], window[1], 1/30000),touch_lick_perch.mean(axis=0),  'C5')
    ax5.plot(np.arange(window[0], window[1], 1/30000),visual_lick_perch.mean(axis=0),  'C5')

    mTouchFR = np.vstack(spike_counts_touch[::-1][touch_lick_ind]).mean(axis = 0)/bin_size
    mVisualFR = np.vstack(spike_counts_vis[::-1][visual_lick_ind]).mean(axis = 0)/bin_size
    semTouchFR = scipy.stats.sem(np.vstack(spike_counts_touch[::-1][touch_lick_ind]))/bin_size
    semVisualFR = scipy.stats.sem(np.vstack(spike_counts_vis[::-1][visual_lick_ind]))/bin_size
    
    mTouchFR_nl = np.vstack(spike_counts_touch[::-1][~touch_lick_ind]).mean(axis = 0)/bin_size
    mVisualFR_nl = np.vstack(spike_counts_vis[::-1][~visual_lick_ind]).mean(axis = 0)/bin_size
    semTouchFR_nl = scipy.stats.sem(np.vstack(spike_counts_touch[::-1][~touch_lick_ind]))/bin_size
    semVisualFR_nl = scipy.stats.sem(np.vstack(spike_counts_vis[::-1][~visual_lick_ind]))/bin_size

    ax6.plot(edges[1:], mTouchFR, 'C2')
    ax7.plot(edges[1:], mVisualFR, 'C2')
    ax6.plot(edges[1:], mTouchFR_nl, 'C7')
    ax7.plot(edges[1:], mVisualFR_nl, 'C7')
    ax6.fill_between(edges[1:], mTouchFR-semTouchFR, mTouchFR+semTouchFR, color ='C2', alpha = 0.5)
    ax7.fill_between(edges[1:], mVisualFR-semVisualFR, mVisualFR+semVisualFR, color = 'C2', alpha = 0.5)
    ax6.fill_between(edges[1:], mTouchFR_nl-semTouchFR_nl, mTouchFR_nl+semTouchFR_nl, color ='C7', alpha = 0.5)
    ax7.fill_between(edges[1:], mVisualFR_nl-semVisualFR_nl, mVisualFR_nl+semVisualFR_nl, color = 'C7', alpha = 0.5)


    for a in axes[2:]:
        a.set_xlim(window)

    for a in [ax, ax2]:
        a.set_ylabel('Trials')

    ax4.set_ylabel('Normalized Perch')
    ax6.set_ylabel('Firing rate (FR)')

    for a in axes:
        a.spines['right'].set_visible(False)
        a.spines['top'].set_visible(False)
        a.xaxis.set_ticks_position('bottom')
        a.yaxis.set_ticks_position('left')

    for ax in plt.gcf().axes:
        try:
            ax.label_outer()
        except:
            pass

    fig.subplots_adjust(wspace = .15)
    cax = fig.add_axes([0.92, 0.66, 0.01, 0.22])
    fig.colorbar(im2,cax=cax)
    
    return fig
```
